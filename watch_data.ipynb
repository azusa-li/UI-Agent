## actionType匹配原数据

```python
file_name_ls = [
    './query_results3/Kyle.json',
    './query_results3/Im.json',
    './query_results3/Cyberpunk.json',
    './query_results3/dcynsd.json',
    './query_results3/Siing.json',
    './query_results3/Kyle..json',
    './query_results3/sk.json',
    './query_results3/cceatmore.json',
    './query_results3/Jingqiu.json',
    './query_results3/HATBTBS.json',
]
```

```python
def show_dict(data_dict):
    print(json.dumps(data_dict, ensure_ascii=False, indent=4))
```

```python
import json

action_dict = {}

for file_name in file_name_ls:
    all_data = json.load(open(file_name, 'r'))
    for data in all_data:
        steps = json.loads(data["steps"])
        for step in steps:
            if "type" in step.keys():
                action_dict[step["type"]] = action_dict.get(step["type"], 0) + 1

# show_dict(action_dict)

total_cnt = 0
for key, value in action_dict.items():
    total_cnt += value

from tabulate import tabulate

sorted_items = sorted(action_dict.items(), key=lambda x: x[1], reverse=True)
table_data = [[key, value, f"{value/total_cnt:.2%}"] for key, value in sorted_items]
headers = ["Action Type", "Count", "Percentage"]
print(tabulate(table_data, headers=headers, tablefmt="grid"))
```

## 查看query_results下每个数据的step情况

```python
dir_path = './query_results3'
import os

for file in os.listdir(dir_path):
    if file.endswith('.json'):
        file_path = os.path.join(dir_path, file)
        with open(file_path, 'r') as f:
            all_data = json.load(f)
            print(f"File: {file}")
            print(f"Number of steps: {len(all_data)}")
            for data in all_data:
                steps = json.loads(data["steps"])
                # print(len(steps))
                for step_num, step in enumerate(steps):
                    if "axTree" in step.keys():
                        print(f"Step {step_num} has axTree!")
                print("-"*5)
    print("*"*10)                
```

## 查看哪些action里面有value 以及有action和有axtree的对应情况

```python
import json
import os

dir_path = './query_results3'
import os

action_without_value = {}
total_cnt = 0

for file in os.listdir(dir_path):
    file_path = os.path.join(dir_path, file)
    with open(file_path, 'r') as f:
        all_data = json.load(f)
        print(f"File: {file}")
        for data_num, data in enumerate(all_data):
            steps = json.loads(data["steps"])
            # print(len(steps))
            for step_num, step in enumerate(steps):
                if "value" in step.keys() and len(step["value"]) == 0:
                    action_type = step["type"]
                    action_without_value[action_type] = action_without_value.get(action_type, 0) + 1
                if "value" not in step.keys():
                    print(file, data_num, step_num)
                total_cnt += 1

print(action_without_value)
print(sum(action_without_value.values())/total_cnt)
```
