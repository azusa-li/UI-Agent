## 查看axtree的节点，value等情况

````python
import json

index_ls = [1, 2, 3, 4, 5]

all_sample_axtree = []
for index in index_ls:
    sample_axtree_path = f"./raw_axtree/selected_50/aliuliuliuliu/{index}/1.json"
    with open(sample_axtree_path, 'r') as f:
        sample_axtree = json.load(f)
    all_sample_axtree.append(sample_axtree)
````

````python
def count_node_roles(node, role_counts=None):
    if role_counts is None:
        role_counts = {}
    
    if node is None:
        return role_counts
    
    # Count current node's role
    role = node.get('role', 'no_role')
    value = node.get('name', '')
    is_value = len(value) > 0
    role_counts[role] = role_counts.get(role, {"count": 0, "is_value": 0})
    role_counts[role]["count"] += 1
    role_counts[role]["is_value"] += is_value

    # Recursively process children
    children = node.get('children', [])
    for child in children:
        count_node_roles(child, role_counts)
    
    return role_counts

# 使用示例:
role_counts = count_node_roles(sample_axtree)

# 打印结果
print("Node role counts:")
for role, count in sorted(role_counts.items()):
    print(f"{role}: {count}")

# 打印有value的节点占比
total_cnt = sum(count["count"] for count in role_counts.values())
value_cnt = sum(count["is_value"] for count in role_counts.values())
print(f"有value的节点占比: {value_cnt/total_cnt:.2%}")


````

````python
role_counts = {}
for sample_axtree in all_sample_axtree:
    role_counts = count_node_roles(sample_axtree, role_counts)

# 打印结果
print("Node role counts:")
for role, count in sorted(role_counts.items()):
    print(f"{role}: {count}")

# 打印有value的节点占比
total_cnt = sum(count["count"] for count in role_counts.values())
value_cnt = sum(count["is_value"] for count in role_counts.values())
print(f"有value的节点占比: {value_cnt/total_cnt:.2%}")
````

## 查看step中的value和axtree节点中name值的对应情况

````python
from utils import *
import numpy as np

step_value_ls = []
axtree_name_ls = []

user_name = "cceatmore"

sample_user_data_path = f"./query_results3/selected_50/{user_name}_selected_50_0_50_new.json"
with open(sample_user_data_path, 'r') as f:
    sample_user_data = json.load(f)

for task_num, task in enumerate(sample_user_data):
    steps = json.loads(task["steps"])
    for step_num, step in enumerate(steps):
        if "axTree" in step and step["axTree"] and "value" in step and step["type"] != "type":
            raw_axtree_path = f"./raw_axtree/selected_50/{user_name}/{task_num}/{step_num}.json"
            with open(raw_axtree_path, 'r') as f:
                raw_axtree = json.load(f)
            path = ["html"] + step["path"].split('>')
            raw_retrieved_axtree = find_node_by_path(raw_axtree, path)
            if raw_retrieved_axtree is None:
                raw_retrieved_axtree = {}
            # print(step["value"])
            # print(raw_retrieved_axtree["name"])
            if step["value"] != raw_retrieved_axtree.get("name", "") and step["value"] != "" and raw_retrieved_axtree.get("name", "") != "":
                print(f"Task Num: {task_num}, Step Num: {step_num}")
                print(f"Value: {step['value']}")
                print(f"Axtree Name: {raw_retrieved_axtree.get('name', '')}")
                print("\n----------------------------------\n")
                print(json.dumps(step, ensure_ascii=False, indent=4), "\n\n")
                print(json.dumps(raw_retrieved_axtree, ensure_ascii=False, indent=4))
                print("\n\n\n", "="*30, "\n\n\n")
            step_value_ls.append(step["value"])
            axtree_name_ls.append(raw_retrieved_axtree.get("name", ""))
            # print("-"*10)
    # break


````

````python
def calculate_similarity(str1, str2):
    """计算两个字符串的相似度，基于编辑距离"""
    from Levenshtein import distance
    max_len = max(len(str1), len(str2))
    if max_len == 0:
        return 1.0
    return 1 - distance(str1, str2) / max_len

# 设置相似度阈值
SIMILARITY_THRESHOLD = 0.8

# 比较step中的value和axtree节点中name值的对应情况
equal_cnt = 0
similar_cnt = 0
total_cnt = 0

# 排除axtree的name为空 排除step value为空
for i in range(len(axtree_name_ls)):
    if axtree_name_ls[i] != "" and step_value_ls[i] != "":
        similarity = calculate_similarity(step_value_ls[i], axtree_name_ls[i])
        if step_value_ls[i] == axtree_name_ls[i]:
            equal_cnt += 1
            similar_cnt += 1
        elif similarity >= SIMILARITY_THRESHOLD:
            similar_cnt += 1
            print(f"相似但不完全相同:")
            print(f"Step value: {step_value_ls[i]}")
            print(f"Axtree name: {axtree_name_ls[i]}")
            print(f"相似度: {similarity:.2%}")
            print("-"*10)
        total_cnt += 1

print(f"完全相等数量: {equal_cnt}")
print(f"相似数量: {similar_cnt}")
print(f"总数量: {total_cnt}")
print(f"完全相等比例: {equal_cnt/total_cnt:.2%}")
print(f"相似比例: {similar_cnt/total_cnt:.2%}")
````
